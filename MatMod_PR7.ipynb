#1. Плоскость 20х40

```python
import numpy as np
from numpy.linalg import norm

K = 5

k = np.array([-K, 0, 0])

k_norm = K / norm(k)



f = open('/content/plate_20_40 без доп инф.dat','r')  # открыть файл из рабочей директории в режиме чтения
lines = []

for line in f.readlines():
  lines.append(line.strip().split(' '))
```

```python
lines = lines[2:-2]
```

```python
lines
for line in lines:
  line = [[line[x], line[x+1], line[x+2]] for x in range(0, len(line), 3)]
```

```python
lines
```

```python
lines_done = [[[float(lst[x]), float(lst[x+1]), float(lst[x+2])] for x in range(0,len(lst),3)] for lst in lines]
```

```python
k_norm
```

```python
#Сетка
lines_done
```

```python
x = []
y = []
z = []

for line in lines_done:
  for dot in line:
    x.append(dot[0])
    y.append(dot[1])
    z.append(dot[2])

len(x)
```

```python
# importing mplot3d toolkits
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
 
fig = plt.figure(figsize=(10, 10))
 
# syntax for 3-D projection
ax = plt.axes(projection ='3d')
 
# defining axes
ax.scatter(z, x, y) 
# syntax for plotting
ax.set_title('PLATE')
plt.show()
```

```python
#Расчет M0

M0 = []
for line in lines_done:
  M0.append((np.array(line[0]) + np.array(line[1]) + np.array(line[2]) + np.array(line[3]))/ 4)


M0 = np.array(M0)
M0
```

```python
#Расчет n0

n0 = []
for line in lines_done:
  AB = (np.array(line[1]) + np.array(line[2]) - np.array(line[0]) - np.array(line[3])) / 2
  CD = (np.array(line[0]) + np.array(line[1]) - np.array(line[2]) - np.array(line[3])) / 2
  n0.append(np.cross(AB, CD, axisa=0, axisb=0))

np.array(n0).shape
```

```python
n0
```

```python
S = []
for line in n0:
  S.append(norm(line))
```

```python
S[1]
```

```python
n = [n / norm(n) for n in n0]
```

```python
n
```

```python
#Расчет матрицы A0

A0 = []
for line in lines_done:
  idx = lines_done.index(line)
  #A -> B
  M2M1 = np.array(line[1]) - np.array(line[0])
  M1M0 = np.array(M0[idx] - line[0])
  M2M0 = np.array(M0[idx] - line[1])
  p = norm(M1M0) ** 2 * norm(M2M1) ** 2 - np.dot(M2M1, M1M0) ** 2
  W1 = np.cross(M2M1, M1M0) * (- (np.dot(M2M1, M2M0) / norm(M2M0)) + np.dot(M2M1, M1M0) / norm(M1M0)) / p
  #B -> C
  M2M1 = np.array(line[2]) - np.array(line[1])
  M1M0 = np.array(M0[idx] - line[1])
  M2M0 = np.array(M0[idx] - line[2])
  p = norm(M1M0) ** 2 * norm(M2M1) ** 2 - np.dot(M2M1, M1M0) ** 2
  W2 = np.cross(M2M1, M1M0) * (- (np.dot(M2M1, M2M0) / norm(M2M0)) + np.dot(M2M1, M1M0) / norm(M1M0)) / p
  #C -> D
  M2M1 = np.array(line[3]) - np.array(line[2])
  M1M0 = np.array(M0[idx] - line[2])
  M2M0 = np.array(M0[idx] - line[3])
  p = norm(M1M0) ** 2 * norm(M2M1) ** 2 - np.dot(M2M1, M1M0) ** 2
  W3 = np.cross(M2M1, M1M0) * (- (np.dot(M2M1, M2M0) / norm(M2M0)) + np.dot(M2M1, M1M0) / norm(M1M0)) / p
  #D -> A
  M2M1 = np.array(line[0]) - np.array(line[3])
  M1M0 = np.array(M0[idx] - line[3])
  M2M0 = np.array(M0[idx] - line[0])
  p = norm(M1M0) ** 2 * norm(M2M1) ** 2 - np.dot(M2M1, M1M0) ** 2
  W4 = np.cross(M2M1, M1M0) * (- (np.dot(M2M1, M2M0) / norm(M2M0)) + np.dot(M2M1, M1M0) / norm(M1M0)) / p

  W = W1 + W2 + W3 + W4

  A0.append(np.dot(W, n[idx]))

print(A0)
```

```python
#Расчет матрицы А

def K1(x, y, nx, ny, k):
  #Приводим к комплексным
  x.astype(complex)
  y.astype(complex)
  nx.astype(complex)
  ny.astype(complex)
  k = complex(k)

  #Считаем r и К1

  r = norm(x - y)

  K1 = (
        ( k**2 * np.exp(-1j * k * r) ) / r**3 -\
        ( 3 * 1j * k * np.exp(-1j * k * r) ) / r**4 -\
        ( 3 * np.exp(-1j * k * r) - 3 ) / r**5
       ) *\
       np.dot((x - y), nx) * np.dot((x - y), ny) +\
       ( np.exp(-1j * k * r) + 1j * k * r * np.exp(-1j * k * r) - 1 ) / r**3 *\
        np.dot(nx, ny)
  return K1

#Создаем матрицу из комплексных единиц, затем заполняем через цикл
size = M0.shape[0]
print(size)
A = np.zeros(shape=[size, size], dtype=complex)
for i in range(size):
  for j in range(size):
    if i != j:
      A[i][j] = K1(M0[i], M0[j], n[i], n[j], K) * S[i]
```

```python
#Расчет правых частей
f = np.zeros(shape=size, dtype=complex)

for i in range(size):
  f[i] = np.dot(n[i], k) * 1j * K * np.exp(-1j * K * np.dot(M0[i], k))
```

```python
#Нахождение матрицы G - решение СЛАУ

A_overall = A0 + A

g = np.linalg.solve(A_overall, f)

G_plate = np.zeros(shape=[size, 2])

for i in range(size):
  G_plate[i][0] = np.real(g[i])
  G_plate[i][1] = np.imag(g[i])

print(G_plate)
```

```python
#Визуализация результата

# importing mplot3d toolkits
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
 
fig = plt.figure(figsize=(10, 10))
 
# syntax for 3-D projection
ax = plt.axes(projection ='3d')
 
# defining axes
ax.scatter(z, x, y, c = np.vstack([np.abs(g), np.abs(g), np.abs(g), np.abs(g)]).reshape(-1), cmap='magma') 
# syntax for plotting
ax.set_title('PLATE')
plt.show()
```

```python
#Интерактивный график
import plotly.express as px
fig = px.scatter_3d(x=x, y=y, z=z,
              color=np.vstack([np.abs(g), np.abs(g), np.abs(g), np.abs(g)]).reshape(-1))
fig.show()
```

```python
#Сохранение файла
FILE = open("result_of_plate_final.dat", "w")
FILE.write(f'{len(G_plate)}\n')

for result in G_plate:
  FILE.write(f'{result[0]} {result[1]}\n')
```
