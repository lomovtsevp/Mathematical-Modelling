```python
import pandas as pd

def credit_calc(SK, PS, KP):
  AP = SK * (PS / 12 + (PS / (12 * ((1 + PS / 12) ** KP - 1))))

  NP0 = SK * PS / 12
  OK0 = SK

  periods = [i for i in range(1, KP + 1)]
  NPs = []
  NPs.append(NP0)
  OKs = []
  OKs.append(OK0)
  OC0 = AP - NP0
  OCs = []
  OCs.append(OC0)

  for i in range(1, KP):
    OKs.append(OKs[i-1] - OCs[i-1])
    NPs.append(OKs[i] * PS / 12)
    OCs.append(AP - NPs[i])

  return pd.DataFrame({'Номер периода' : periods,
                       'ОК' : OKs,
                       'НП' : NPs,
                       'ОС' : OCs})
```

```python
credit_calc(1000000, 0.2, 19)
```
